Se procede a la importación de las librerías 

```python
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import Select, WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
from datetime import datetime, timedelta
import pandas as pd
import requests
import urllib.request, urllib.parse, urllib.error 
import datetime
import time
```

Como se va a estudiar los datos del mes de octubre para ello generamos una lista:

```python
inicio = datetime.datetime(2020,10,1)
fin    = datetime.datetime(2020,10,30)

fechas = [(inicio + timedelta(days=d)).strftime("%d-%m-%Y")
                    for d in range((fin - inicio).days + 1)] 
```

Se crea una lista con las páginas del mes para consultar:

```python
pagebase="https://www.esios.ree.es/es/generacion-y-consumo?date="
page=[]
for i in range(len(fechas)):
    page.append(pagebase + fechas[i])
```

Se crea un dataframe para guardar los valores de las diferentes fechas:

```python
BalEnergEspOct20 = pd.DataFrame([],columns =["Fecha","Potencia Disponible","Potencia Indisponible","Potencia Generada Instalada",
                                        "Potencia Generada sin CO2 diaria","Potencia Eolica diaria","Potencia Hidraulica diaria",
                                        "Potencia Solar diaria","Potencia Nuclear diaria","Demanda Real diaria",
                                        "Demanda Programada diaria","Demanda Prevista diaria","Generación diaria",
                                        "Interconexiones diarias","Consumo diario"])
```

Se activa el driver de selenium para poder evaluar el código de la página:

```python
driver = webdriver.Firefox()
wait = WebDriverWait(driver,timeout=30)
```

```python
for i in range(len(page)):
    #Se crea una variable para almacenar los valores temporales de cada fila
    va=[]
    #Se consigue el codigo de la pagina selecionada en el valor i
    driver.get(page[i])
    #Se introduce un tiempo para que se cargue la página
    time.sleep(20)
    #Se usa la variable soup para tratar el código obtenido de la página
    soup = BeautifulSoup(driver.page_source, features='lxml')
    #Se obtienen los diferentes valores necesarios
    valores = soup.findAll("div", {"class": "is-number"})
    valores1 = soup.findAll("div", {"class": "is-medium-number"})
    valores2 = soup.findAll("div", {"class": "is-small-number"})
    #Se almacenan los valores según como se ha definido las columnas del dataframe
    va = [fechas[i],valores[1].text,valores[2].text,valores[0].text,valores1[0].text,valores2[0].text,
          valores2[1].text,valores2[2].text,valores2[3].text,valores[11].text,valores[12].text,valores[13].text,
          valores2[4].text,valores2[6].text,valores2[8].text]
    #Se almacenan los valores en el dataframe
    BalEnergEspOct20.loc[i+1]=va
```

Se para el driver

```python
driver.quit()
```
